```python
plt.figure(figsize = (15,15))

plt.plot(Oil_Prod, label = Oil_Prod.columns)

plt.legend(loc = 'best')

plt.title('Oil_Production of Every Countries',size = 15, color = 'Blue')
```

```python
columns = Oil_Prod.columns
```

```python
scaled_Oil_Prod = Oil_Prod
for i in columns:
    scaled_Oil_Prod[i] = Oil_Prod[i]/Oil_Prod[i].max()
    
    
```

```python
plt.figure(figsize=(8,4))
plt.plot(scaled_Oil_Prod['India'])
plt.plot(scaled_Oil_Prod['Total World '])
plt.figure(figsize=(8,4))
plt.plot(scaled_Oil_Prod['India'])
plt.plot(scaled_Oil_Prod['China'])
plt.figure(figsize=(8,4))
plt.plot(scaled_Oil_Prod['India'])
plt.plot(scaled_Oil_Prod['OECD Americas'])
```

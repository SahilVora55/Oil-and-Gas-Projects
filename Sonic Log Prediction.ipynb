### Scaling the Data

```python
scaler.fit(df11B)
scaled_df11B = scaler.transform(df11B)
```

```python
scaler.fit(df11A)
scaled_df11A = scaler.transform(df11A)
```

```python
scaled_df1A = pd.DataFrame(scaled_df1A, columns = ['CALI', 'DRHO', 'DT','GR','NPHI','PEF','RHOB','ROP','RT'])
scaled_df1B = pd.DataFrame(scaled_df1B, columns = ['CALI', 'DRHO', 'DT','GR','NPHI','PEF','RHOB','ROP','RT'])
scaled_df1C = pd.DataFrame(scaled_df1C, columns = ['CALI', 'DRHO','GR','NPHI','PEF','RHOB','ROP','RT'])
scaled_df11A = pd.DataFrame(scaled_df11A, columns = ['CALI', 'DRHO', 'DT','GR','NPHI','PEF','RHOB','ROP','RT'])
scaled_df11B = pd.DataFrame(scaled_df11B, columns = ['CALI', 'DRHO','GR','NPHI','PEF','RHOB','ROP','RT'])
```

```python
scaled_df1A.fillna(0, inplace=True)
scaled_df1B.fillna(0, inplace=True)
scaled_df1C.fillna(0, inplace=True)
scaled_df11A.fillna(0, inplace=True)
scaled_df11B.fillna(0, inplace=True)
```

### Take 1A and 1B as training dataset and 11A as testing dataset

```python
df_train = scaled_df1A.append(scaled_df1B)
df_test = scaled_df11A
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
X1 = df_train.drop('DT',axis=1)
y1 = df_train['DT']
```

```python
model = RandomForestRegressor()
```

```python
model.fit(X1,y1)
```

### Prediction of Test data of Well 11A and check model accuaracy.

```python
scaled_df11A['DT_pred'] = model.predict(df_test.drop('DT',axis=1))
```

```python
plt.figure(figsize = (5,24))

plt.plot(scaled_df11A['DT_pred'],df11A.index)
plt.plot(scaled_df11A['DT'],df11A.index)
```

```python
from sklearn.metrics import mean_squared_error as mse
```

```python
mse(scaled_df11A['DT_pred'],scaled_df11A['DT'])
```

```python
from sklearn.metrics import mean_absolute_error as mae
```

```python
mae(scaled_df11A['DT_pred'],scaled_df11A['DT'])
```

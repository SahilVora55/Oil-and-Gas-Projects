```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Hydrostatic Pressure at bottom of the Well

```python
well_head_P = 100  # psi
Depth = 6000       # ft
SG_wat = 1.07
SG_oil = 0.8
SG_gas = 0.15

Hydro_grad_wat = 0.433*SG_wat
Hydro_grad_oil = 0.433*SG_oil
Hydro_grad_gas = 0.433*SG_gas

FBHP_wat = Hydro_grad_wat*Depth + well_head_P
FBHP_oil = Hydro_grad_oil*Depth + well_head_P
FBHP_gas = Hydro_grad_gas*Depth + well_head_P
```

```python
FBHP_gas
```

```python
wat_points = np.linspace(well_head_P,FBHP_wat,20)
Depth_points = np.linspace(0,Depth,20)
oil_points = np.linspace(well_head_P,FBHP_oil,20)
gas_points = np.linspace(well_head_P,FBHP_gas,20)
```

```python

plt.figure(figsize=(8,6))
plt.title('Hydrostatic Pressure vs Depth',size= 14, c='red')
plt.plot(wat_points,Depth_points,label = 'Salt Water')
plt.plot(oil_points,Depth_points,label = 'Oil')
plt.plot(gas_points,Depth_points, label = 'Gas')
plt.ylim(6000,0)
plt.legend()
plt.grid(alpha = 0.4)
plt.xlabel('FBHP',size=12,color='orange')
plt.ylabel('Depth',size=12,c='orange')
```

```python
#  1000 ft water, 3000 ft oil, 2000 ft gas
```

```python
# without injection gas
FBHP1 = 1000*Hydro_grad_wat + 3000*Hydro_grad_oil + 2000*Hydro_grad_gas + well_head_P
FBHP1
```

```python
# After injection at 2000 ft depth
Depth_inj = 2000
Hydro_grad_inj = 0.15  # above inj point
FBHP2 = (1000*Hydro_grad_wat + 3000*Hydro_grad_oil + 2000*Hydro_grad_gas)*((Depth-Depth_inj)/Depth) + well_head_P + Hydro_grad_inj*Depth_inj
FBHP2
```

```python
above_inj_depth = np.linspace(0,Depth_inj,20)
inj__press_grad = np.linspace(well_head_P,Hydro_grad_inj*Depth_inj,20)
below_inj_depth = np.linspace(Depth_inj,Depth,20)
original_press_grad = np.linspace(Hydro_grad_inj*Depth_inj,FBHP2, 20)

without_inj_press = np.linspace(well_head_P,FBHP1,20)
```

```python

plt.figure(figsize=(8,6))
plt.title('FBHP_after_gas_inj. vs Depth',size=14,color='blue')
plt.plot(inj__press_grad,above_inj_depth,color='green',label='Inj. at 2000 ft')
plt.plot(original_press_grad,below_inj_depth,color='green')
plt.plot(without_inj_press,Depth_points,color='red',label='before Inj.')
plt.xlabel('FBHP',size=12,color='blue')
plt.ylabel('Depth',size=12,color='blue')
plt.axhline(2000,label='Inj. Depth',linestyle='--',alpha=0.5)
plt.legend()
plt.ylim(6000,0)
plt.xlim(0,2500)
plt.grid(alpha=0.4)
```

```python
# After injection at 4000 ft depth
Depth_inj2 = 4000
Hydro_grad_inj = 0.15  # above inj point
FBHP2 = (1000*Hydro_grad_wat + 3000*Hydro_grad_oil + 2000*Hydro_grad_gas)*((Depth-Depth_inj2)/Depth) + well_head_P + Hydro_grad_inj*Depth_inj2
FBHP2
```

```python
above_inj_depth2 = np.linspace(0,Depth_inj2,20)
inj__press_grad2 = np.linspace(well_head_P,Hydro_grad_inj*Depth_inj2,20)
below_inj_depth2 = np.linspace(Depth_inj2,Depth,20)
original_press_grad2 = np.linspace(Hydro_grad_inj*Depth_inj2,FBHP2, 20)

without_inj_press = np.linspace(well_head_P,FBHP1,20)
```

```python

plt.figure(figsize=(8,6))
plt.title('FBHP_after_gas_inj. vs Depth',size=14,color='blue')

plt.plot(inj__press_grad,above_inj_depth,color='green',label='Inj. at 4000 ft')
plt.plot(original_press_grad,below_inj_depth,color='green')

plt.plot(without_inj_press,Depth_points,color='red',label='before Inj.')
plt.plot(inj__press_grad2,above_inj_depth2,label='Inj. at 2000 ft',c='b')

plt.plot(original_press_grad2,below_inj_depth2,c='b')
plt.axhline(2000,label='Inj. Depth',linestyle='--',alpha=0.5)

plt.xlabel('FBHP',size=12,color='blue')
plt.ylabel('Depth',size=12,color='blue')

plt.axhline(4000,label='Inj. Depth',linestyle='--',alpha=0.5)
plt.legend()

plt.ylim(6000,0)
plt.xlim(0,2500)

plt.grid(alpha=0.4)
```

# Subplot:- Multiple plots next to each other

```python
fig , axs = plt.subplots(nrows=2, ncols=2)

fig.set_figheight(6) ; fig.set_figwidth(10)


x_data = np.linspace(1,100,1000)

axs[0,0].plot(x_data, x_data**1, label='Power = 1')
axs[0,0].set_title('Power = 1')

axs[0,1].plot(x_data, x_data**2, label='Power = 2', color='red')
axs[0,1].set_title('Power = 2')

axs[1,0].plot(x_data, x_data**3, label='Power = 3', color='brown')
axs[1,0].set_title('Power = 3')

axs[1,1].plot(x_data, x_data**4, label='Power = 4', color='orange')
axs[1,1].set_title('Power = 4')

fig.tight_layout()
```

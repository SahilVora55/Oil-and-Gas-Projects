```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

### import data

```python


df = pd.read_csv('https://raw.githubusercontent.com/Divyanshu-ISM/Machine-Learning-Deep-Learning/main/Volve_PF14.csv')
```

```python
df.head()
```

```python
import datetime
```

```python
df['DATEPRD'] = pd.to_datetime(df['DATEPRD'])
```

```python
df.set_index('DATEPRD',inplace = True)
```

```python
df.head()
```

### fill nan value with zero

```python
df = df.fillna(0)  
```

```python
df.describe().T
```

```python
df.columns
```

### Ploted Average downhole Pressure and WI rate in one plot with subplot and twinx()

```python
plt.rcParams["figure.autolayout"] = True

ax1 = plt.subplot()
l1, = ax1.plot(df.index,df['AVG_DOWNHOLE_PRESSURE'], color='green')

ax2 = ax1.twinx()

l2, = ax2.plot(df.index,df['BORE_WI_VOL'], alpha = 0.3,color='blue')

plt.legend([l1, l2], ["AVG_DOWNHOLE_PRESSURE", "BORE_WI_VOL"])

plt.show()
```

### replace zero values with some values between IQR or median 

```python
df['AVG_DOWNHOLE_PRESSURE'].replace(0,np.random.randint(240,260),inplace = True)
```

```python
df['BORE_WI_VOL'].replace(0,np.random.randint(4500,6500),inplace = True)
df['BORE_WAT_VOL'].replace(0,df['BORE_WAT_VOL'].median(),inplace = True)
df['BORE_OIL_VOL'].replace(0,df['BORE_OIL_VOL'].median(),inplace = True)
```

```python
df.replace(0,df.median(),inplace = True)
```

# Create ML model to predict GLR

```python
df['LIQ_VOL']  =   df['BORE_OIL_VOL']   + df['BORE_WAT_VOL']
```

```python
sns.scatterplot(df['LIQ_VOL'],df['AVG_DOWNHOLE_PRESSURE'])
```

```python
X_train = df[['LIQ_VOL']][:2000]
y_train = df[['AVG_DOWNHOLE_PRESSURE']][:2000]
X_test = df[['LIQ_VOL']][2000:]
y_test = df[['AVG_DOWNHOLE_PRESSURE']][2000:]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
yp_test = model.predict(X_test)
```

```python
plt.scatter(X_train, y_train )
plt.scatter(X_test,yp_test)
```

# Predicting the Oil and Water Production rate using ML algorithms

```python
fig, ax = plt.subplots(2, 1, sharex='col', sharey='row')



ax[0].plot(df.index,df['BORE_OIL_VOL'],color = 'orange')
ax[1].plot(df.index,df['BORE_WAT_VOL'],color = 'blue')

```

### split data into training and testing sets

```python
# for oil

X1_train = df[['AVG_DOWNHOLE_PRESSURE','BORE_GAS_VOL']][:2000]
y1_train = df[['BORE_OIL_VOL']][:2000]
X1_test = df[['AVG_DOWNHOLE_PRESSURE','BORE_GAS_VOL']][2000:]
y1_test = df[['BORE_OIL_VOL']][2000:]
```

```python
# for water 

X2_train = df[['AVG_DOWNHOLE_PRESSURE','BORE_WI_VOL','BORE_GAS_VOL']][:2000]
y2_train = df[['BORE_WAT_VOL']][:2000]
X2_test = df[['AVG_DOWNHOLE_PRESSURE','BORE_WI_VOL','BORE_GAS_VOL']][2000:]
y2_test = df[['BORE_WAT_VOL']][2000:]
```

### apply linear regression

```python
from sklearn.linear_model import LinearRegression
```

```python
model1 = LinearRegression()
```

```python
model2 = LinearRegression()
```

```python
model1.fit(X1_train,y1_train)
```

```python
model2.fit(X2_train,y2_train)
```

```python
yp1 = model1.predict(X1_test)
```

```python
yp2 = model2.predict(X2_test)
```

### Visualizing our result

```python
fig, ax = plt.subplots(2, 1, sharex='col', sharey='row')

                            
ax[0].plot(df.index[:2000],y1_train,color = 'orange',label = 'Past Oil Prod Rate')
ax[0].plot(df.index[2000:],yp1,color = 'red',label = 'Predicted Oil Prod Rate')

xmin, xmax = ax[0].get_xlim()
ax[0].set_xticks(np.round(np.linspace(xmin, xmax, 5), 2))

ax[1].plot(df.index[:2000],y2_train,color = 'blue',label = 'Past Water Prod Rate')
ax[1].plot(df.index[2000:],yp2,color = 'red',label = 'Predicted Water Prod Rate')

plt.xlabel('DATEPRD', color = 'brown', size = '14')
plt.ylabel('Production Rate',va = 'center', color = 'brown',size = '14')
ax[0].legend()
ax[1].legend()
fig.tight_layout()
```
